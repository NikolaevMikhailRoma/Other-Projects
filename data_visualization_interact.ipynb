```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from ipywidgets import interact
import seaborn as sns

def download_file(path):
    df = pd.read_csv(path, skiprows =2, sep = '\t', decimal =  ',', encoding='cp1251')
    df['delta_t'] = delta_t(df)
    df['t'] = [convert_to_seconds(df['Время'][i]) for i in range(df['Время'].shape[0])]
    return df

def convert_to_seconds(time):
    t = int(time[:2]) * 3600 + int(time[3:5]) * 60 + float(time[6:])
    return t

def delta_t(df, column = 'Время'):
    delta_t = [convert_to_seconds(df[column][i+1]) - convert_to_seconds(df[column][i]) for i in range(df[column].shape[0]-1)]
    delta_t.append(0.05)
    return delta_t

def plot_interact_times():
    @interact(start_index=(0, df.shape[0], 1), lenght=(0, df.shape[0], 5))

    def plot_f(start_index, lenght):
        fig, ax = plt.subplots(figsize = (10, 7), nrows=2)
        ax[0].plot(df['delta_t'][start_index:start_index+lenght])
        ax[0].set(title='file_path : {}'.format(path), ylabel='delta_t (Time_n+1 - Time_n) (sec)')

        ax[1].plot(df['t'][start_index:start_index+lenght])
        ax[1].set(xlabel='Index', ylabel='time after midnight (sec)')

        plt.show()
        
def plot_interact_column():

    @interact(start_time=(int(df['t'].min()), int(df['t'].max()), 1), 
              number_of_indices=(0, df.shape[0], 5),
              column=list(df.columns[4:]),
             )

    def plot_f(start_time, number_of_indices, column):
        fig, ax = plt.subplots(figsize = (10, 5))
        data = df[df['t'] > start_time][:number_of_indices]
        ax.plot('t', column, data = data)
        ax.scatter('t', column, data = data, s = 10, c = 'green')
        ax.set(title='file_path : {}'.format(path), xlabel='Time (s)', ylabel='value')
    #     ax.legend()
        plt.show()
        
def plot_interact_bar_time():
    @interact(max_delta_time = (0, int(df['delta_t'].max()) + 1, 0.05))
    def plot_f(max_delta_time = 0.4):
        fig, ax = plt.subplots(figsize = (10, 5))
        data = df[df['delta_t'] <= max_delta_time]
        data['delta_t'] = [round(i, 2) for i in data['delta_t']]
        bins = len(data['delta_t'].unique())     
        ax.hist(data['delta_t'], bins = bins-1, align = 'left', rwidth = 0.5)
        ax.set_xlim(0, data['delta_t'].max())
        plt.show()
```
